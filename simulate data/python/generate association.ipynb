```python
import numpy as np
import pickle
import os
import pandas as pds
import datetime
```

```python
name_DF = 'DF_N=44138_2023-01-05.csv'
DF = pds.read_csv(os.path.join('..', 'datasets', name_DF), delimiter = ',')
DF = DF[~DF.duplicated()] # delete duplicates
DF = DF.dropna() # delete NaN values
DF
```

```python
with open(os.path.join('..', 'dictionaries', 'country_proba_names_75names.pkl'), 'rb') as f:
    country_proba_names = pickle.load(f)
```

```python
# generate treatment
DF['treatment'] = np.random.binomial(n = 1, p = 0.4, size = DF.shape[0])

# generate covariates
DF['X1'] = 2020 - DF['birth year'] # age
DF['X2'] = np.random.normal(loc = np.random.randint(1,11), scale = np.random.uniform(0,2), size = DF.shape[0])
DF['X3'] = np.random.normal(loc = np.random.randint(1,11), scale = np.random.uniform(0,2), size = DF.shape[0])
DF['X4'] = np.random.normal(loc = np.random.randint(5,11), scale = np.random.uniform(0,0.5), size = DF.shape[0])
DF['X5'] = np.random.normal(loc = np.random.randint(1,11), scale = np.random.uniform(0,2), size = DF.shape[0])

# generate outcome
residual_errors = np.random.normal(size = DF.shape[0])
a = np.random.randint(1,21)/10
b = np.random.randint(1,31)/10
c = np.random.randint(1,41)/10
d = np.random.randint(1,11)/10
DF['Y'] = - 10 + a*DF['treatment']*DF['X1'] + b*np.log(DF['X4']) + c*DF['X3']*DF['X2'] + d*DF['X5']

# ATE: a * Expectation(X1) 
```
